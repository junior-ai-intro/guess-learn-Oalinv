**11. A cookie tomorrow v. today**

```python
import numpy as np
from maze import Maze

maze = Maze()
q = np.zeros((4,4,4))
solved = False
while not solved:
    state = maze.reset()
    done = False
    while not done:
        action = maze.sample_n()                         
        new_state, reward, done = maze.step(action)
        q[state[0]][state[1]][action] += reward
        if max(new_state) < 4 and min(new_state) >= 0:
            q[state[0]][state[1]][action] += 0.9 * max(q[new_state[0]][new_state[1]])
        state = new_state
        solved = True if max(q[0][0]) > 0 else False

Maze.plot(q,f=max)
print(q)
```
